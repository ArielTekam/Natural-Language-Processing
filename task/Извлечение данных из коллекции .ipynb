```python
def parse_line(line):
    parts = line.strip().split('\t')
    if len(parts) != 3:
        return None
    topic, title, text = parts
    return {
        "topic": topic,
        "title": title,
        "text": text
    }

parsed_data = [parse_line(l) for l in lines]
parsed_data = [p for p in parsed_data if p is not None]

print(f"{len(parsed_data)} lignes valides.")
parsed_data[0]
```

```python
def find_first_article_with_person(data, max_docs=500):
    for i, item in enumerate(data[:max_docs]):
        text = item["text"]
        doc = Doc(text)
        doc.segment(segmenter)
        doc.tag_morph(morph_tagger)
        doc.tag_ner(ner_tagger)

        persons = [span for span in doc.spans if span.type == "PER"]
        if persons:
            # нормализуем имена
            for p in persons:
                p.normalize(morph)
            return i, item, persons
    return None, None, []

index, article, persons = find_first_article_with_person(parsed_data, max_docs=500)

print("Индекс статьи:", index)
print("Тема:", article["topic"])
print("Заголовок:", article["title"])
print("\nТекст (первые 300 символов):")
print(article["text"][:300], "...\n")

print("Найденные персоны:")
for p in persons:
    print("→", p.text, "| normalized:", p.normal)
```

```python
from dataclasses import dataclass
from typing import Optional, List

@dataclass
class Entry:
    # Имя персоны
    name: str
    # Дата рождения (если найдена, иначе None)
    birth_date: Optional[str]
    # Место рождения (если найдено, иначе None)
    birth_place: Optional[str]
```

```python
import re

# Регулярные выражения для поиска даты
# Полная дата типа "12 мая 1978 года"
full_date_pattern = re.compile(
    r'(\d{1,2}\s+[А-Яа-яЁё]+(?:я|а|е|у|ом)?\s+\d{4}\s+года?)'
)

# Год рождения типа "1987 года" или "1987 год"
year_pattern = re.compile(
    r'(\d{4})\s+года?'
)

birth_keywords = [
    'родился',
    'родилась',
    'уроженец',
    'уроженка',
    'родом из'
]


def extract_entries_from_text(text: str) -> List[Entry]:
    """
    Простая эвристическая функция:
    1) находит предложения с ключевыми словами о рождении
    2) внутри них ищет персоны (PER), локации (LOC/ORG) и даты
    3) для каждой персоны создаёт Entry
    """
    entries: List[Entry] = []

    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.tag_ner(ner_tagger)

    # нормализуем все сущности заранее
    for span in doc.spans:
        span.normalize(morph)

    # проходим по всем предложениям
    for sent in doc.sents:
        sent_text = sent.text
        sent_lower = sent_text.lower()

        # пропускаем предложения без ключевых слов
        if not any(kw in sent_lower for kw in birth_keywords):
            continue

        # находим все сущности в текущем предложении
        sent_spans = [
            span for span in doc.spans
            if span.start >= sent.start and span.stop <= sent.stop
        ]

        persons = [s for s in sent_spans if s.type == 'PER']
        locations = [s for s in sent_spans if s.type in ('LOC', 'ORG')]

        # ищем дату в тексте предложения
        birth_date = None
        m_full = full_date_pattern.search(sent_text)
        if m_full:
            birth_date = m_full.group(1)
        else:
            m_year = year_pattern.search(sent_text)
            if m_year:
                birth_date = m_year.group(1)

        # выбираем первое место как место рождения (эвристика)
        birth_place = locations[0].text if locations else None

        # создаём записи для всех персон в этом предложении
        for p in persons:
            entry = Entry(
                name=p.normal,          # нормализованное имя
                birth_date=birth_date,  # строка с датой или None
                birth_place=birth_place # строка с местом или None
            )
            entries.append(entry)

    return entries
```

```python
# Протестируем функцию на первых 100 текстах
all_entries: List[Entry] = []

for item in parsed_data[:100]:
    text = item["text"]
    entries = extract_entries_from_text(text)
    all_entries.extend(entries)

print("Найдено записей:", len(all_entries))
for e in all_entries[:10]:
    print(e)
```

```python
import re
from typing import List

year_pattern = re.compile(r'(\d{4})')  # просто любая 4-значная дата

def extract_entries_from_text_v2(text: str) -> List[Entry]:
    """
    Эвристика попроще:
    - ищем предложения, где одновременно есть:
      * хотя бы одна персона (PER)
      * хотя бы одна 4-значная дата (год)
      * хотя бы одна локация (LOC или ORG)
    - создаём Entry для каждой персоны
    """
    entries: List[Entry] = []

    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.tag_ner(ner_tagger)

    for span in doc.spans:
        span.normalize(morph)

    for sent in doc.sents:
        sent_text = sent.text

        # сущности внутри предложения
        sent_spans = [
            span for span in doc.spans
            if span.start >= sent.start and span.stop <= sent.stop
        ]

        persons = [s for s in sent_spans if s.type == 'PER']
        locations = [s for s in sent_spans if s.type in ('LOC', 'ORG')]

        if not persons:
            continue

        m_year = year_pattern.search(sent_text)
        if not m_year:
            continue

        birth_date = m_year.group(1)
        birth_place = locations[0].text if locations else None

        for p in persons:
            entry = Entry(
                name=p.normal,
                birth_date=birth_date,
                birth_place=birth_place
            )
            entries.append(entry)

    return entries
```

```python
all_entries_v2 = []

for item in parsed_data[:300]:  # протестировать на первых 300 товарах
    text = item["text"]
    entries = extract_entries_from_text_v2(text)
    all_entries_v2.extend(entries)

print("Найдено записей:", len(all_entries_v2))
for e in all_entries_v2[:10]:
    print(e)
```

```python
all_entries_v2 = []

for item in parsed_data:
    text = item["text"]
    entries = extract_entries_from_text_v2(text)
    all_entries_v2.extend(entries)

print("Всего найдено записей:", len(all_entries_v2))
```

```python
import pandas as pd

# Преобразуем список Entry в DataFrame
df_entries = pd.DataFrame([vars(e) for e in all_entries_v2])

print("Размер таблицы:", df_entries.shape)
df_entries.head()
```
